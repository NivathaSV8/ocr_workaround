```python
from google.colab import drive
drive.mount('/content/drive')
img_path = '/content/drive/My Drive/form_sample.jpg'
```

```python
from paddleocr import TextRecognition
model = TextRecognition(model_name="PP-OCRv5_server_rec")
output = model.predict(input='/content/drive/My Drive/form_sample.jpg', batch_size=1)

for res in output:
    res.print()
    res.save_to_img(save_path="/content/drive/My Drive/output/")
    res.save_to_json(save_path="/content/drive/My Drive/output/res.json")
```

```python
from paddleocr import PaddleOCR

ocr = PaddleOCR(
    lang="japan",
    ocr_version="PP-OCRv5",
    text_recognition_model_name="PP-OCRv5_mobile_rec",
)
result = ocr.predict('/content/drive/My Drive/form_sample.jpg')

for res in result:
    res.print()
    res.save_to_img("/content/drive/My Drive/output/")
    res.save_to_json("/content/drive/My Drive/output/")
```

```python
from paddleocr import TableRecognitionPipelineV2

pipeline = TableRecognitionPipelineV2()
output = pipeline.predict('/content/drive/My Drive/input/routetable.jpg')
for res in output:
    # res.print() ## Print the predicted structured output
    # res.save_to_xlsx("/content/drive/My Drive/output/routetable.xlsx")
    res.save_to_xlsx("/content/drive/My Drive/output/")
    res.save_to_img("/content/drive/My Drive/output/")
    res.save_to_json("/content/drive/My Drive/output/")
```

```python
import cv2
import matplotlib.pyplot as plt

def display_image(image_path, name):
  img = cv2.imread(image_path)

  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  # Display the image using Matplotlib
  plt.imshow(img_rgb)
  plt.title(name)
  plt.axis('off')  # Hide axes for a cleaner image display
  # plt.figure().set_figwidth(15)
  plt.show()

display_image("/content/drive/My Drive/output/routetable_layout_det_res.jpg", "Det")
display_image("/content/drive/My Drive/output/routetable_preprocessed_img.jpg", "PreProcessed")
display_image("/content/drive/My Drive/output/routetable_ocr_res_img.jpg", "OCRed")
```
